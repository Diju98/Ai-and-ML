```python
from sklearn.cluster import Birch
```

```python
df.columns.values
df2=df.drop('Area_Name', axis=1)
df2=df2.drop('Subgroup', axis=1)
```

```python
df2.head()
```

```python
model=Birch(branching_factor=30, n_clusters=5, threshold=2.5 )
model.fit(df2)
```

```python
prediction=model.predict(df2)
print(prediction)
```

```python
plt.scatter(df2["Victims_Between_10-14_Yrs"],df2["Victims_Between_14-18_Yrs"], c=prediction, cmap='Paired',alpha=0.5)
plt.show()
```

```python
plt.scatter(df2["Victims_Between_18-30_Yrs"],df2["Victims_Between_30-50_Yrs"], c=prediction, cmap='Paired',alpha=0.5)
plt.show()
```
